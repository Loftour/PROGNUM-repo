```python
import numpy as np
from numpy import linalg as la
np.set_printoptions(suppress=True)

A = np.array([[1, 2, 3],[0, -4, 1],[0, 3, -1]])

det=la.det(A)
print(det)

b=np.array([2, 2, -1])
xA_b=np.linalg.solve(A, b)
print(f'(x,y,z) is {xA_b}')
is_orthogonal = np.allclose(A.T, la.inv(A))
print(is_orthogonal)

# A*v=lambda*v, therefore la.eig(a)=lambda + eig vectors
Lambda, v = la.eig(A)
print(Lambda)
print(v)
```

```python
E=np.array([[-0.06698874, -0.87275577, -0.48353891],
    [ 0.49272847, -0.45034696,  0.74458463],
    [-0.86760081, -0.1883746 ,  0.46019978]])
is_orthogonal=np.allclose(E.T,la.inv(E)) #compared E.T*E to identity matrix
print(is_orthogonal)

det=np.linalg.det(E) #det of 1 to ensure rotation and not reflection
print(det)
print('as the matrix have a determinant of near one and is orthogonal it is a rotation matrix')

Lambda=la.eig(E)
print(Lambda)
print('as one of the vectors is 1 it reinforces the noticion of a rotation matrix')
```

```python
import matplotlib.pyplot as plt
x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])

def solve_interpolation(x, y):
    n=len(x)
    V=np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            V[i, j]=x[i]**j
    coeffs=np.linalg.solve(V, y)
    return coeffs

coeffs=solve_interpolation(x, y)
x_plot=np.linspace(min(x), max(x), 500)
y_plot=np.zeros_like(x_plot)
for i in range(len(coeffs)):
    y_plot += coeffs[i] * (x_plot**i)

plt.scatter(x,y)
plt.plot(x_plot,y_plot)
print(coeffs)
#check this over
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Define the data
x_data = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y_data = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])

class interpolation:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.f_linear = interp1d(x, y, kind='linear')
        self.poly_coeffs = np.polyfit(x, y, len(x) - 1)
        self.f_poly = np.poly1d(self.poly_coeffs)
        self.f_spline = interp1d(x, y, kind='cubic')


    def calculate(self, x_val, method='spline'):
        if method == 'linear':
            return self.f_linear(x_val)
        elif method == 'polynomial':
            return self.f_poly(x_val)
        elif method == 'spline':
            return self.f_spline(x_val)
        else:
            return "Invalid method"

interp_obj=interpolation(x_data, y_data)

X=5.0
print(f"Values at x = {X}:")
print(f"Linear: {interp_obj.calculate(X, 'linear'):.4f}")
print(f"Polynomial: {interp_obj.calculate(X, 'polynomial'):.4f}")
print(f"Spline: {interp_obj.calculate(X, 'spline'):.4f}")

x_fine=np.linspace(x_data.min(), x_data.max(), 100)
plt.scatter(x_data, y_data, color='red', label='Data Points')
plt.plot(x_fine, interp_obj.f_spline(x_fine), label='Cubic Spline', color='blue')
plt.title('Cubic Spline Interpolation')
plt.legend()
plt.show()
```
