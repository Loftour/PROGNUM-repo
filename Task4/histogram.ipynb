```python
x=np.random.normal(185,8,10000)
import matplotlib.pyplot as plt
import scipy.stats as scipy
s=np.std(x)
m=np.mean(x)
plt.hist(x,40, color="red")
plt.axvline(m,color="blue",label="mean height")
plt.axvspan(m-s,m+s,alpha=0.4,label="1 standard deviation")
plt.legend()
print(np.std(x))
print(np.mean(x))
cdf = scipy.norm.cdf(195,loc=185, scale=1000)
P=1-cdf
P *= 100        # to percentage
print(f"Probability to find a height of 195 cm or lower in our sample is: {P:.2f} %")
std = scipy.norm.cdf(m+s,loc=m-s, scale=2*s)
print(f"Probability to find a height between 1 standard deviation is: {std:.2f} %")
```
